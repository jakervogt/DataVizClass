## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df=purchase_data.drop_duplicates(subset='Item Name', keep="first")
items_df
```

```python
popular_item = items_df[["Item ID", "Item Name","Price"]]
popular_item.reset_index(drop=True)
```

```python
number_ofitems=purchase_data.groupby(["Item Name"]).count()
```

```python
mergedpopitems=pd.merge(popular_item,number_ofitems, on="Item Name")
dropped_items=mergedpopitems.drop(['SN','Age','Gender','Item ID_y','Price_y','Group'], axis = 1)
dropped_items
```

```python
totalpurchase_value=dropped_items["Price_x"]*dropped_items["Purchase ID"]
totalpurchase_value
```

```python
totalpurchase_value_df=pd.DataFrame(totalpurchase_value)
totalpurchase_value_df.insert(0, 'Item ID_x', dropped_items["Item ID_x"])
totalpurchase_value_df
```

```python
all_merged=pd.merge(dropped_items,totalpurchase_value_df, on="Item ID_x")
all_merged
all_merged_descending=all_merged.sort_values(by='Purchase ID', ascending=False)
all_merged_descending.columns = ['Item ID', 'Item Name', 'Price', 'Number Purchased', 'Total Amount']
finished_merged_descending= all_merged_descending[['Item ID', 'Item Name','Number Purchased','Price','Total Amount']]
finished_merged_descending
```

## Most Profitable Items

```python
most_profitable=all_merged_descending.sort_values(by="Total Amount",ascending=False)
```
